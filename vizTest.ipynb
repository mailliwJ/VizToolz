```python
import numpy as np
import pandas as pd

import viztools as viz
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv('./data/titanic.csv')
titanic.head()
```

```python
viz.plotAll4(titanic, cat_col='class', num_col='fare', show_mean=True, kde=True,hist_color='blue', alpha=0.4)
```

```python
viz.categorical_relationships(titanic, cat_col1='sex', cat_col2='class', relative_freq=True, show_values=True)
```
